```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_folder_path = "../../dataset"
if os.path.exists(dataset_folder_path):
    review_files = os.listdir(dataset_folder_path)

dataset2 = pd.DataFrame()

for file_name in review_files:
    file = pd.read_csv(f"{dataset_folder_path}/{file_name}",sep=',', header=0)
    random_sample = file.sample(n=3000)
    dataset2 = dataset2.append(random_sample)
    
print(dataset2.shape)
```

```python
dataset2 = dataset2["review_text"]
```
